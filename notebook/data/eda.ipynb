```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("stud.csv")
```

```python
df.head()
```

```python
df["average"]=(df['math_score']+df["reading_score"]+df["writing_score"])/3
```

```python
df.head()
```

```python
df["race_ethnicity"].unique()
```

```python
df.nunique()
```

```python
mathscore_lessthan_20=df[df['math_score']<=20]['average'].count()
readingscore_lessthan_20=df[df['reading_score']<=20]['average'].count()
writingscore_lessthan_20=df[df['writing_score']<=20]['average'].count()
print('the no of students having less than 20 in maths is',mathscore_lessthan_20)
print('the no of students having less than 20 in reading is',readingscore_lessthan_20)
print('the no of students having less than 20 in writing is',writingscore_lessthan_20)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='average',hue='gender',bins=30,kde=True)
```

##females are tend to get more marks compared to boys


```python
sns.catplot(data=df,x='average',y='race_ethnicity',hue='gender',kind='box')
```

```python
race_counts = df['race_ethnicity'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
race_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Race/Ethnicity')
plt.ylabel('')  # Hides the y-axis label
plt.show()
```
